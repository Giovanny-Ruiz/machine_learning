### In Synthetic data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
````

````python
# Generar datos de clasificación
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_informative=2, n_clusters_per_class=1, n_classes=2, random_state=42)

# Crear una figura y un eje
_, ax = plt.subplots(figsize=(8, 6))

# Graficar los datos
ax.scatter(X[:, 0], X[:, 1], c=y, cmap='Paired')
plt.show()
````

````python
# Create an instance
decision_tree = DecisionTreeClassifier()

# Train the model
decision_tree.fit ( X , y ) #(# Samples, # Features)
````

### **Decision boundaries**

````python
# Min-Max values for the grid of decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

# Create grid
x_grid, y_grid = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))

# Estimate predictions over the grid
predictions = decision_tree.predict(np.c_[x_grid.ravel(), y_grid.ravel()])
print(predictions.shape)
````

````python
# Reshape predictions
predictions = predictions.reshape(x_grid.shape)

# Draw grid
_, ax = plt.subplots(figsize=(8, 5))
ax.pcolormesh(x_grid, y_grid, predictions, cmap="Paired")

# Opcional: Añadir los puntos de datos originales
ax.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap="Paired")
plt.show()
````

### In Real data

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_iris
````

````python
# Cargar el conjunto de datos de iris
iris = load_iris()
iris_x = iris.data  # Características
iris_y = iris.target  # Etiquetas

# Imprimir formas
print (iris_x.shape, iris_y.shape)
````

````python
# Crear una instancia del clasificador
decision_tree = DecisionTreeClassifier()

# Entrenar el modelo
decision_tree.fit(iris_x, iris_y)  # (muestras, características)
````

````python
# Crear una figura y un eje
_, ax = plt.subplots(figsize=(12, 8))

# Graficar el árbol de decisión
plot_tree(decision_tree, filled=True, fontsize=5, ax=ax)
plt.show()
````
