# Let's start on the basics. Variables and arrays

### First, import some libraries

```
import numpy as np
```

```
N = 19
array_1 = np.arange(N) # Create an array of N values
array_2 = np.random.randint(0, 100, size=N) # Create an array of random int values
            # with the same lenght as array_1

print("array_1:", array_1)
print("array_2:", array_2) # print both array_1 and array_2
print("Shape of array_1:", array_1.shape)
print("Shape of array_2:", array_2.shape) # print both array_1 and array_2 shapes
print("Summation of array_1 and array_2:", array_1 + array_2) # print summation of array_1 and array_2
```

```
array_1[3] = 70 # Replace the value of array_1 in pos 3 by 70
print("Updated summation of array_1 and array_2:", array_1 + array_2) # Print summation again
```

```
elementwise_multiplication = array_1 * array_2 # Apply element-wise multiplication among array_1 and array_2
print("Element-wise multiplication of array_1 and array_2:", elementwise_multiplication)# Print results
```

```
# Import some libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

## Load an image using OpenCV

```
# Loading a color image
img = cv2.imread('/content/drive/MyDrive/Machine/machine_learning/Sem_01/lena.jpg' )
# Convert BGR image to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) # plots an image
plt.axis('off')  # Hide axes
plt.show()
```

### Let's play with the data

```
# Printing the image shape (size of the image)
print("Image shape:", img.shape)
```

```
# Looking through the image...
for i in range(0, img.shape[0], 100):  # Step by 100 rows
    for j in range(0, img.shape[1], 100):  # Step by 100 columns
        print(f"Pixel value at ({i}, {j}): {img_rgb[i, j]}")
# print among image indexes
plt.imshow(img_rgb)
plt.axis('off')  # Hide axes
plt.show()
```

```
# Changing its channels values!
img = cv2.imread( '/content/drive/MyDrive/Machine/machine_learning/Sem_01/lena.jpg' )
img[:, :, [0, 2]] = img[:, :, [2, 0]]
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) # Voila!
plt.axis('off')  # Hide axes
plt.show()
```

### Going further

```
# Messing everything around.
img = cv2.imread('/content/drive/MyDrive/Machine/machine_learning/Sem_01/lena.jpg')

# Convert BGR image to RGB for displaying
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot the modified image
plt.imshow(img_rgb)
plt.axis('off')  # Hide axes
plt.show()

# Loading the image in grayscale
img_gray = cv2.imread('/content/drive/MyDrive/Machine/machine_learning/Sem_01/lena.jpg', cv2.IMREAD_GRAYSCALE)

# Plot the grayscale image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```

```
# Playing with image intensities
img_gray = cv2.imread('/content/drive/MyDrive/Machine/machine_learning/Sem_01/lena.jpg', cv2.IMREAD_GRAYSCALE)

for r in range(img_gray.shape[0]):  # Go through rows
    for c in range(img_gray.shape[1]):  # Go through cols
        if img_gray[r, c] > 128:  # Create a simple condition
            img_gray[r, c] = 255  # Do something (e.g., set to white)
        else:
            img_gray[r, c] = 0  # Do something (e.g., set to black)

# Plot the modified grayscale image
plt.imshow(img_gray, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```
